```python
import scipy as sp
relPath = "/backup/dev/ipython/build_machine_learning_system_with_python/data/ch01/data/"
dataPath = relPath + "web_traffic.tsv"
print dataPath
data = sp.genfromtxt(dataPath,delimiter = "\t")
print(data[:10])
```

```python
print(data.shape)
```

scipy에서 데이터를 2개의 벡터로 나눔

첫번째 벡터 x는 시간을 나타냄.

두번째 벡터 y는 특정 시간의 요청 수를 

나누는 방법 : 개별적으로 열을 선택할 수 있는 scipy의 특수 인덱스 표기를 사용 

```python
x = data[:,0]
y = data[:,1]
```

y 값 중에 일부 값이 잘못 된 값 확인 

```python
sp.sum(sp.isnan(y))
```

```python
sp.sum(sp.isnan(x))
```

```python
x = x[~sp.isnan(y)]
y = y[~sp.isnan(y)]
```

데이터 정제 

데이터 형태를 파악하기 위해 matplotlib를 이용하여 점도표(scatter) 표시 

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.title("web traffic over the last month")
plt.xlabel("Time")
plt.ylabel("Hits/hour")
plt.xticks([w*7*24 for w in range(10)],['week %i for w in range(10)'])
plt.autoscale(tight=True)
plt.grid()
plt.show()     
```
