# 4. 주제 모델링

#### 3장 복습
- 3장에서는 텍스트를 그룹으로 군집화. 3장의 방식은 매우 유용하나 항상 적합하지는 않다.
- 예) 이 책의 주제는 파이썬 or 머신러닝? 답) 두 주제 모두 정답이다. 그러나 3장에서는 이렇게 처리할 수 없다.

#### 4장 주제
- 4장의 목표는 주제 모델링(Topic modeling)을 배우는 것.
- 군집화를 이용하지 않고, 주제(Topic)라 불리는 작은 그룹 수로 구분하는 기법을 배운다.
- 또한 텍스트의 중심부에 있는 주제와 모호하게 이야기된 텍스트 사이를 어떻게 유추하는지도 배운다.


### 잠재 디리클레 할당(Latent Dirichlet Allocation, 이하 LDA)

#### 개요
- Unsupervised Generative Topic Model
- 모든 문서는 하나의 토픽에만 속하는 것이 아니라 두세가지의 다른 토픽들의 혼합(Mixture)으로 정의한다는 가설

#### LDA와 생성 모델(Generative Model)
<img src="images/lda_model2.png" width="606" />
- '생성 모델(Generative Model)'은 어떤 확률분포와 그 파라미터가 있다고 할 때, 그로부터 랜덤 프로세스에 따라 데이터를 생성하는 모델.
- 즉, 문서의 주제 분포와 각 주제별로 특정 단어를 생성할 확률을 알고 있으면, 특정 문서가 만들어질 확률을 계산할 수 있다는 이야기.


### 문서(Document)-주제(Topic) 분포, 주제(Topic)-단어(Term) 분포
<img src="images/lda_description.png" width="725" />
- 단어와 주제의 분포를 알고 있으면(Term Matrix), 어떤 단어를 많이 쓸수록 주제에 가까워지는지 알 수 있다.
- 이를 통해 문서에서 어느 단어를 많이 쓰고 있는지 계산해서, 단어에 맞는 주제를 찾아낸다.
- 이러한 방식을 통해 문서의 주제를 파악할 수 있다.

### 실습 1: 주제 모델 만들기

````python
# -*- coding: utf-8 -*-
from __future__ import unicode_literals
from gensim import corpora, models, similarities
import os 
os.chdir("/backup/dev/ipython/build_machine_learning_system_with_python/4장/") 

# 연합 통신사(AP, Associated Press)의 뉴스 리포트 데이터셋을 읽어온다.
# http://www.cs.princeton.edu/~blei/lda-c/ap.tgz
corpus = corpora.BleiCorpus('./data/ap/ap.dat', './data/ap/vocab.txt')

# 모델을 생성한다.
model = models.ldamodel.LdaModel(corpus, num_topics=100, id2word=corpus.id2word, alpha=None)

# 이를 통해 주제(Topic) 모델을 만든다.
topics = [model[c] for c in corpus]

# 출력해본다. 출력은 짝 리스트, 즉 (topic_index, topic_weight) 이다.
print(topics[0])
````

### 실습 2: 주제 개수의 히스토그램 만들기(alpha=None)

````python
# 주제 개수의 히스토그램을 그릴 수 있다.

import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image 

# 그래프에 출력할 수 있도록 문서/주제를 정리한다.
num_topics_used1 = [len(model[doc]) for doc in corpus]

# 그래프를 그린다. X축은 주제의 개수, Y축은 문서의 개수.
plt.hist(num_topics_used1, np.arange(42))
plt.ylabel('Number of documents')
plt.xlabel('Number of topics')
plt.savefig('images/Figure_04_01.png')
plt.clf()

# 그래프를 출력한다.
Image(filename='images/Figure_04_01.png')
````

### 실습 3: 주제 개수의 히스토그램 만들기(alpha=1.0)

````python
# 좀 더 넓히기 위해, 함수의 매개변수인 알파(alpha)를 사용한다.
# 알파의 정확한 의미는 약간 추상적이지만, 알파값이 클수록 더 많은 주제로 이어진다.

# 알파값을 1.0으로 변경한 후 모델을 재생성한다.
model2 = models.ldamodel.LdaModel(corpus, num_topics=100, id2word=corpus.id2word, alpha=1.0)

# 그래프에 출력할 수 있도록 문서/주제를 정리한다.
num_topics_used2 = [len(model2[doc]) for doc in corpus]

# 그래프를 그린다. 변함없이 X축은 주제의 개수, Y축은 문서의 개수.
plt.hist([num_topics_used1, num_topics_used2], np.arange(42))
plt.ylabel('Number of Documents')
plt.xlabel('Number of Topics')

# 기본 알파(alpha=None)와 1.0 알파(alpha=1.0)의 차이점을 보여준다.
plt.text(9, 223, 'alpha=None')
plt.text(26, 156, 'alpha=1.0')
plt.savefig('images/Figure_04_02.png')
plt.clf()

# 그래프를 출력한다.
Image(filename='images/Figure_04_02.png')
````

### 실습4: 주제를 활용해 워드 클라우드(Word Cloud) 만들기

````python
from os import path
from wordcloud import WordCloud

# Word Cloud를 만드는 함수
def create_cloud(filename, words):
    wordcloud = WordCloud(font_path='/Users/shaynekang/Library/Fonts/malgun.ttf')
    wordcloud.generate(words)
    wordcloud.to_file(filename)

# First, we need to sum up the weights across all the documents
weight = np.zeros(model.num_topics)

for doc in corpus:
    for col, val in model[doc]:
        weight[col] += val
        # As a reasonable alternative, we could have used the log of val:
        # weight[col] += np.log(val)
max_topic = weight.argmax()

# 상위 64개의 주제를 꺼내온다.
words = model.show_topic(max_topic, 64)

# 워드 클라우드를 만들 수 있도록 list를 string으로 변경(구분자는 \n으로)
words = '\n'.join(map(lambda w: w[1], words))

# 워드 클라우드를 생성한다.
create_cloud('images/cloud_blei_lda.png', words)

# 워드 클라우드를 출력한다.
Image(filename='images/cloud_blei_lda.png')
````

### 실습5: 주제를 활용한 유사도 비교

````python
import numpy as np
from scipy.spatial import distance
        
# 주제 매트릭스를 만든다. 가로는 주제의 개수, 세로는 문서의 개수
dense = np.zeros((len(topics), 100), float)
for ti, t in enumerate(topics):
    for tj, v in t:
        dense[ti, tj] = v

# 주제 매트릭스를 활용해 거리 매트릭스를 생성한다. 모든 짝의 거리를 계산한다.
pairwise = distance.squareform(distance.pdist(dense))

# 편법: 대각선에 가장 높은 값을 설정한다.
largest = pairwise.max()
for ti in range(len(topics)):
    pairwise[ti, ti] = largest + 1

# 가장 유사도가 높은 문서를 찾는다.
def closest_to(doc_id):
    return pairwise[doc_id].argmin()

document = closest_to(2)
print(document)
````

### 실습 6: 위키피디아 모델링

````python
import logging
import gensim
import numpy as np

id2word = gensim.corpora.Dictionary.load_from_text('data/wikipedia/wiki_en_output_wordids.txt.bz2')
mm = gensim.corpora.MmCorpus('data/wikipedia/wiki_en_output_tfidf.mm')

model = gensim.models.ldamodel.LdaModel.load('data/wikipedia/wiki_lda.pkl')

topics = []
for doc in mm:
    topics.append(model[doc])
    
lens = np.array([len(t) for t in topics])

# 평균 문서는 6.3개의 주제를 다룸.
print(np.mean(lens))

# 94%는 10개보다 적은 주제를 다룬다.
print(np.mean(lens <= 10))
````

### 실습7: 위키피디아에서 가장 많이 사용한 단어(w/ 워드 클라우드)

````python
counts = np.zeros(100)

for doc_top in topics:
        for ti, _ in doc_top:
            counts[ti] += 1
            
words = model.show_topic(counts.argmax(), 64)
words = '\n'.join(map(lambda w: w[1], words))

create_cloud('images/wikipedia_most.png', words)

# 워드 클라우드를 출력한다.
Image(filename='images/wikipedia_most.png')
````

### 실습8: 위키피디아에서 가장 적게 사용한 단어(w/ 워드 클라우드)

````python
words = model.show_topic(counts.argmin(), 64)
words = '\n'.join(map(lambda w: w[1], words))

create_cloud('images/wikipedia_least.png', words)

# 워드 클라우드를 출력한다.
Image(filename='images/wikipedia_least.png')
````
